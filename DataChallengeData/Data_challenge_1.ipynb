# Feature engineering

## Data cleaning

```python
df = df.dropna()
```

```python
df.info()
```

## Handling Airline

```python
df = pd.concat([df, pd.get_dummies(df['Airline'])], axis=1)
df.drop(['Airline'], axis=1, inplace=True)
df
```

## Handling Date of journey
Since it's a periodic feature, I am going to convert it into cyclical representation using sine function

```python
df['Date_of_Journey'] = pd.to_datetime(df['Date_of_Journey'])
df
```

```python
df['Date_of_Journey'] = df['Date_of_Journey'].dt.dayofyear
df
```

```python
import numpy as np
df['Date_of_Journey'] = np.sin(2 * np.pi * df['Date_of_Journey'] / 365)
df
```

## Handling source and destination

```python
df = pd.get_dummies(df, columns=['Source', 'Destination'])
df
```

## Handling route and total stops
Since route is hard to be converted into numeric and it largely related to total stop, I want to use only total stops instead

```python
df.drop('Route', axis=1, inplace=True)
df
```

```python
to_numeric_total_stops = {'non-stop' : 0, '1 stop' : 1, '2 stops' : 2, '3 stops' : 3, '4 stops' : 4}
df.replace(to_numeric_total_stops, inplace=True)
df['Total_Stops'].unique()
```

## Handling departure time and arrival time
They are periodic features so I'm going to convert them using sine function

```python
df['Arrival_Time'] = pd.to_datetime(df['Arrival_Time'])
df
```

```python
df['Arrival_Time'].unique()
```

```python
df['Arrival_Time'] = np.sin(2 * np.pi * (df['Arrival_Time'].dt.hour * 60 + df['Arrival_Time'].dt.minute) / 1440)
df
```

```python
df['Dep_Time'] = pd.to_datetime(df['Dep_Time'])
df['Dep_Time'] = np.sin(2 * np.pi * (df['Dep_Time'].dt.hour * 60 + df['Dep_Time'].dt.minute) / 1440)
df
```

## Handling additional info

```python
df = pd.concat([df, pd.get_dummies(df['Additional_Info'])], axis=1)
df.drop(['Additional_Info'], axis=1, inplace=True)
df
```

# Model and Evaluate

```python
from sklearn.model_selection import train_test_split

train_x, test_x, train_y, test_y = train_test_split(df.drop(['Normalized_Price'], axis=1), df['Normalized_Price'], test_size=0.2)
train_x
```

```python
test_x
```

```python
train_y
```

```python
test_y
```

## Linear regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

reg = LinearRegression()
reg.fit(train_x, train_y)
np.sqrt(mean_squared_error(test_y, reg.predict(test_x)))
```

## Ridge
Tunning hyperparameter using grid search

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV

ridge = Ridge()
hyperparameters = {'alpha' : [1, 10, 100], 'max_iter' : [100, 500, 1000]}
ridgeRegressor = GridSearchCV(ridge, hyperparameters, scoring='neg_mean_squared_error', cv=5)
ridgeRegressor.fit(df.drop(['Normalized_Price'], axis=1), df['Normalized_Price'])
print("Best values for hyperparameters : ",ridgeRegressor.best_params_)
```

```python
ridge = Ridge(alpha=1.0, max_iter=100)
ridge.fit(train_x, train_y)
np.sqrt(mean_squared_error(test_y, ridge.predict(test_x)))
```

## Lasso
Tuning hyperparameters using grid search

```python
from sklearn.linear_model import Lasso

lasso = Lasso()
lassoRegressor = GridSearchCV(lasso, hyperparameters, scoring='neg_mean_squared_error', cv=5)
lassoRegressor.fit(df.drop(['Normalized_Price'], axis=1), df['Normalized_Price'])
print("Best values for hyperparameters : ", lassoRegressor.best_params_)
```

```python
lasso = Lasso(alpha=1.0, max_iter=100)
lasso.fit(train_x, train_y)
np.sqrt(mean_squared_error(test_y, lasso.predict(test_x)))
```

Ridge has a better performance of RMSE of 3.46 among these three model. Therefore, I'm going to apply Ridge on the prediction
